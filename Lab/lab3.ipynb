```python
import spacy
from spacy import displacy

nlp = spacy.load('../resume_ner_model/model-best')
```

```python
import os
import fitz

def extract_text_from_pdf(filename):
    pdf = fitz.open(filename)
    cv_text = ""
    
    for page in pdf:
        cv_text += page.get_text() + " "
        
    return cv_text
    

def clean_whitespace(statement):
    """
    Remove any consecutive whitespace characters from the statement text.
    """

    import re 
    
    # Replace linebreaks and tabs with spaces
    statement = statement.replace('_', ' ')
    statement = ' '.join(statement.split())

    # Remove any leading or trailing whitespace
    statement = statement.strip()

    # Remove consecutive spaces
    statement = re.sub(' +', ' ', statement)

    return statement

def clean_text(text):
    import re 
    from pyvi import ViUtils
    text = re.sub(r"[^'/\\&@.+#*%\w\d\s]", ' ', text)
    #remove accents
    text = ViUtils.remove_accents(text).decode('utf-8')
    
    text = clean_whitespace(text)
    
    
    return text

cv = extract_text_from_pdf('../Test_CV/Nguyễn Minh Quân - CV - English.pdf')
cv = clean_text(cv)
```

```python
NER_LABEL = {'PERSON_NAME': 'rgb(238,179,252, 1);',
 'ADDRESS': 'rgb(238,174,202);',
 'EDUCATION': '#FFF5EE',
 'GPA': '#F9F1F0;',
 'SKILL': '#EF9',
 'EXPERIENCE_LEVEL': '#F8AFA6',
 'JOB_TITLE': '#FAEBD7',
 'DATE_BIRTH': '#FFDEAD',
 'MAJOR': '#FFC0CB',
 'MARIAGE_STATUS': '#FFF0F5',
 'ORGANIZATION': '#E0FFFF',
 'GENDER': '#E0FFAA',
 'LOCATION': '#FFAAFF'}
```

```python
options = {'colors': NER_LABEL}
doc_cv = nlp(cv)
displacy.render(doc_cv, style='ent', jupyter=True, options=options)
```
