```python
class PDFChecker(object):
    def __init__(self):
        self.basic_threshold = 0.048882
        self.work_threshold = 0.238021
        self.project_threshold = 0.000398
        self.education_threshold = 0.026845
        self.momentum_threshold = 2

    def detect(self, resume_content:str, resume_heading:dict) -> bool:
        cv_len = len(resume_content)

        if cv_len == 0: return True
        
        basic_threshold = len(resume_heading.get("BASIC", "a"))/cv_len
        work_threshold = len(resume_heading.get("WORK_EXPERIENCE", "a"))/cv_len
        project_threshold = len(resume_heading.get("PROJECT", "a"))/cv_len
        education_threshold = len(resume_heading.get("EDUCATION", "a"))/cv_len
        

        momentum = 0

        momentum += basic_threshold < self.basic_threshold
        momentum += work_threshold < self.work_threshold
        momentum += project_threshold < self.project_threshold
        momentum += education_threshold < self.education_threshold

        return momentum > self.momentum_threshold
```

```python
pdf_checker = PDFChecker()
detect_cv = []
for path in CV_PATHS[:1000]:
    with open(path, 'rb') as f:
        cv_content = pdf_reader.read(f)
        heading_extractor.fit(cv_content)
        heading_dict = heading_extractor.get_dict()
        
        if pdf_checker.detect(cv_content, heading_dict):
            detect_cv.append(heading_dict)

print(len(detect_cv))
        
```

```python
for cv in detect_cv:
    for k, v in cv.items():
        print(f"Key -> {k}: size -> {len(v)} \n{v[:50]}")
    
    print("--------------------------------")
```
